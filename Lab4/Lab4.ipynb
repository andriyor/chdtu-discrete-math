```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.complete_graph(5)
T = nx.minimum_spanning_tree(G)
nx.draw_spring(G, with_labels=True, font_weight='bold')
sorted(T.edges(data=True))
```

```python
G = nx.cycle_graph(4)
G.add_edge(0, 3, weight=2)
T = nx.minimum_spanning_tree(G)
nx.draw_spring(G, with_labels=True, font_weight='bold')
sorted(T.edges(data=True))
```
