```python
import networkx as nx
from networkx.algorithms.flow import edmonds_karp
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
G.add_edge('V1','V2', capacity=12.0)
G.add_edge('V1','V3', capacity=10.0)
G.add_edge('V1','V4', capacity=4.0)
G.add_edge('V2','V4', capacity=8.0)
G.add_edge('V2','V3', capacity=9.0)
G.add_edge('V2','V7', capacity=12.0)
G.add_edge('V3','V5', capacity=4.0)
G.add_edge('V3','V6', capacity=18.0)
G.add_edge('V4','V5', capacity=6.0)
G.add_edge('V5','V7', capacity=15.0)
G.add_edge('V6','V5', capacity=1.0)
G.add_edge('V6','V7', capacity=27.0)
R = edmonds_karp(G, 'V1', 'V7')
flow_value = nx.maximum_flow_value(G, 'V1', 'V7')
nx.draw_circular(G, with_labels=True, font_weight='bold')
print(flow_value)
print(flow_value == R.graph['flow_value'])
```
